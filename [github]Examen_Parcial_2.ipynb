### <font color='blue'>P2.- Realiza una descripción del Clasificador Bayes ingenuo (Naïve Bayes) y proporciona un breve ejemplo ya sea escrito o en código</font>
(Valor: 3 puntos)

---

Respuesta:

*Es una técnica de clasificación supervisada basada en el teorema de Bayes que asume que existe una independencia entre los atributos. Es decir, asumen que la  presencia de una característica particular en una clase no está relacionada con la presencia de cualquier otra característica. Se le llama ineguo ya que asumir independencia absoluta entre todos los atributos, es fácil de modelar y útil para grandes conjuntos de datos.*


```python
#Ejemplo de clasificador Naive Bayes con Iris dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#Importar el conjunto de datos Iris
iris = pd.read_csv('https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/0e7a9b0a5d22642a06d3d5b9bcbad9890c8ee534/iris.csv')
iris.head()
```

```python
#Extracción de columnas con características de largo y ancho del sepalo y pétalo
X = iris.iloc[:,:4]
y = iris['species'].values
print(X.head())
print(y)
```

```python
from sklearn.model_selection import train_test_split
#Fragmentación de dataset a los subconjuntos train y test en 80% y 20% respectivamente
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
#Escalamiento de las características hacia un menor rango para los subconjuntos
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB
#Definición de clasificador Naive Bayes con parámetros predeterminados
classifier = GaussianNB()
#Entrenamiento del clasificador con el subconjunto train
classifier.fit(X_train, y_train)
```

```python
#predicción parael subconjunto test mediante el clasificador Naive Bayes
y_pred = classifier.predict(X_test) 
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
from sklearn.metrics import accuracy_score 
print ("Accuracy : ", accuracy_score(y_test, y_pred))
print('Confusion matrix:\n',cm)
```

```python
# Realizaremos el split de 70/30 para el train y test, respectivamente
#sklearn.model_selection.train_test_split(*arrays, **options)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,stratify = y)
print(X_train,y_train)
#completa
```

```python
# Realizaremos un escalamiento de los datos al conjunto de entrenamiento (esta ya estaba así xd)
sc = StandardScaler()
sc.fit(X_train)

```

```python
# Y deberemos aplicarlo a los conjuntos de X (train y test)

#completa

x_train_std = sc.transform(X_train)



# Apply the SAME scaler to the X test data
x_test_std = sc.transform(X_test)

#completa
```

```python
# Un perceptrón con las siguientes características y deberán entrenarlo
ppn = Perceptron(eta0=0.01, random_state=5)
ppn.fit(x_train_std, y_train)
#completa
```

```python
# Ahora, debemos hacer la predicción...
#completa
y_pred = ppn.predict(x_test_std)
```

```python
# View the predicted y test data
#completa
print(y_pred)
print(y_test) #pretty good xd
```

### P1.- ¿Qué es una serie de tiempo?

Es una sucesión de datos medidos en determinados momentos y ordenados cronológicamente.

### P2.- ¿Por qué y dónde se utiliza la serie de tiempo?

Generalmente se utilizan para construir predicciones, ya que nos permiten observar el comportamiento de los datos a través del tiempo
por lo que da lugar a la posibilidad de detectar patrones en los mismos y de esta manera suponer un comportamiento futuro, aunque
sus aplicaciones pueden ser distintas, simplemente pueden ser para obtener conclusiones sobre algún fenómeno en los datos históricos,
dicho de otro modo, para describir los comportamientos de los datos a través del tiempo. 

Formalizando, el principal objetivo de analizar series de tiempo es desarrollar modelos
que provean descripciones plausibles de la información o dataset. 

### P3.- ¿Cuáles son los componentes de las series de tiempo?

**Nivel:** El valor base para las series si fuese una línea recta.

**Tendencia:** El posible comportamiento de las series a través del tiempo, generalmente es incremental o decremental.

**Estacionalidad:** Los posibles patrones o ciclos que presenta el comportamiento a través del tiempo.   

**Ruido:** La posible varación en las observaciones que son inexplicables para el modelo.

## Predecirán el precio de las bitcoins por meses...
### Verificación de estacionariedad y descomposición STL de la serie

```python
plt.figure(figsize=[15,7])
sm.tsa.seasonal_decompose(df_month.Weighted_Price).plot()
print("Dickey–Fuller test: p=%f" % sm.tsa.stattools.adfuller(df_month.Weighted_Price)[1])
#completa
plt.show()
```

```python
df_month['Weighted_Price_box'], lmbda = stats.boxcox(df_month.Weighted_Price)
print("Dickey–Fuller test: p=%f" % sm.tsa.stattools.adfuller(df_month.Weighted_Price)[1])
```

```python
df_month['prices_box_diff'] = df_month.Weighted_Price_box - df_month.Weighted_Price_box.shift(12)
print("Dickey–Fuller test: p=%f" % sm.tsa.stattools.adfuller(df_month.prices_box_diff[12:])[1])
```

```python
# Regular differentiation
df_month['prices_box_diff2'] = df_month.prices_box_diff - df_month.prices_box_diff.shift(1)
plt.figure(figsize=(15,7))

# STL-decomposition
sm.tsa.seasonal_decompose(df_month.prices_box_diff2[13:]).plot()   
print("Dickey–Fuller test: p=%f" % sm.tsa.stattools.adfuller(df_month.prices_box_diff2[13:])[1])

plt.show()
```

```python
# Initial approximation of parameters using Autocorrelation and Partial Autocorrelation Plots
plt.figure(figsize=(15,7))
ax = plt.subplot(211)
sm.graphics.tsa.plot_acf(df_month.prices_box_diff2[13:].values.squeeze(), lags=45, ax=ax)
ax = plt.subplot(212)
sm.graphics.tsa.plot_pacf(df_month.prices_box_diff2[13:].values.squeeze(), lags=45, ax=ax)
plt.tight_layout()
plt.show()
```

```python
# Initial approximation of parameters
Qs = range(0, 2)
qs = range(0, 3)
Ps = range(0, 3)
ps = range(0, 3)
D=1
d=1
parameters = product(ps, qs, Ps, Qs)
parameters_list = list(parameters)
len(parameters_list)

# Model Selection
results = []
best_aic = float("inf")
warnings.filterwarnings('ignore')
for param in parameters_list:
    try:
        model=sm.tsa.statespace.SARIMAX(df_month.Weighted_Price_box, order=(param[0], d, param[1]), 
                                        seasonal_order=(param[2], D, param[3], 12)).fit(disp=-1)
    except ValueError:
        print('wrong parameters:', param)
        continue
    aic = model.aic
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param, model.aic])
```

```python
# Best Models
result_table = pd.DataFrame(results)
result_table.columns = ['parameters', 'aic']
print(result_table.sort_values(by = 'aic', ascending=True).head())
print(best_model.summary())
```

### P4.- ¿Qué es ADF_Test, ACF, PACF, ARIMA y SARIMA? ¿Para qué se usan?

*    ADF_Test (Augmented Dickey Fuller Test):

    Es una prueba que se le hace a la serie de datos de tiempo para ver si es estacionaria o no, es decir que propiedades como la media y la desviación estandar no varien con el tiempo

*    ACF (AutoCorrelation Function):

    Se usa para ver la correlación entre un dato temporal y sus antecesores

*    PACF (Partial ACF):

    A diferencia de ACF, PACF se usa para ver la correlación entre 2 puntos en el tiempo, sin conciderar las correlaciones de otros puntos en el tiempo

*    ARIMA (Auto Regresive Inregrated Moving Avarage):
    
    Se utiliza para transformar una serie de tiempo cuyo promedio cambia con el tiempo de forma lineal. La serie de tiempo que mos da es una con un promedio constante y así se pueden utilizar otros algoritmos de predicción

*    SARIMA (Seasonal ARIMA): 

    Cuando tienes un patrón que se repite, por ejemplo las estaciones del año, pero la serie de tiempo va aumentando su promedio de forma linea, (como cuando usas ARIMA), es momento de usar SARIMA.



```python
# STL-decomposition
plt.figure(figsize=(15,7))
plt.subplot(211)
best_model.resid[13:].plot()
plt.ylabel(u'Residuals')
ax = plt.subplot(212)
sm.graphics.tsa.plot_acf(best_model.resid[13:].values.squeeze(), lags=48, ax=ax)

print("Dickey–Fuller test:: p=%f" % sm.tsa.stattools.adfuller(best_model.resid[13:])[1])

plt.tight_layout()
plt.show()
```

### Predicciones...

```python
# Inverse Box-Cox Transformation Function
def invboxcox(y,lmbda):
    if lmbda == 0:
        return(np.exp(y))
    else:
        return(np.exp(np.log(lmbda*y+1)/lmbda))
```

```python
# Prediction
df_month2 = df_month[['Weighted_Price']]

# Crea una lista con fechas y que sea hasta el mes de mayo
date_list = [datetime(2020, 11, 30), datetime(2020, 12, 31), datetime(2021, 1, 31), datetime(2021, 2, 28), 
             datetime(2021, 3, 31), datetime(2021, 4, 30), datetime(2021, 5, 31)]


             
future = pd.DataFrame(index=date_list, columns= df_month.columns)
df_month2 = pd.concat([df_month2, future])
df_month2['forecast'] = invboxcox(best_model.predict(start=0, end=125), lmbda)
plt.figure(figsize=(15,7))

#df_month2['Weighted_Price'].plot()
df_month2.Weighted_Price.plot()



df_month2.forecast.plot(color='r', ls='--', label='Predicted Weighted_Price')
plt.legend()
plt.title('Bitcoin exchanges, by months')
plt.ylabel('mean USD')
plt.show()
```
